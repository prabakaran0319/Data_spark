```python
import pandas as pd
import pymysql

# Database connection parameters
host = 'localhost'
user = 'root'
password = 'praba12'
database = 'data_spark'

# Establishing the connection
connection = pymysql.connect(
    host=host,
    user=user,
    password=password,
    database=database
)
cursor = connection.cursor()
print("Database connection established successfully!")

# Table creation queries and CSV file details
tables_and_files = [
    {
        "table_name": "customers",
        "columns": """CustomerKey INT, Gender VARCHAR(50), Name VARCHAR(255), 
                      City VARCHAR(255), State_Code VARCHAR(28), State VARCHAR(255), 
                      Zip_Code VARCHAR(20), Country VARCHAR(255), Continent VARCHAR(255), 
                      Birthday DATE""",
        "file_path": r'C:\Users\praba\Desktop\dataspark_project\df_files\df1_customers.csv',
        "encoding": "latin"
    },
    {
        "table_name": "exchangerates",
        "columns": """Date DATE, Currency VARCHAR(50), Exchange FLOAT""",
        "file_path": r'C:\Users\praba\Desktop\dataspark_project\df_files\df3_exchange_rates.csv',
        "encoding": "latin"
    },
    {
        "table_name": "products",
        "columns": """ProductKey INT, Product_Name VARCHAR(255), Brand VARCHAR(255), 
                      Color VARCHAR(50), Unit_Cost_USD FLOAT, Unit_Price_USD FLOAT, 
                      SubcategoryKey INT, Subcategory VARCHAR(255), CategoryKey INT, 
                      Category VARCHAR(255)""",
        "file_path": r'C:\Users\praba\Desktop\dataspark_project\df_files\df4_products.csv',
        "encoding": "latin"
    },
    {
        "table_name": "sales",
        "columns": """Order_Number INT, Line_Item INT, Order_Date DATE, 
                      CustomerKey INT, StoreKey INT, ProductKey INT, 
                      Quantity INT, Currency_Code VARCHAR(10)""",
        "file_path": r'C:\Users\praba\Desktop\dataspark_project\df_files\df5_sales.csv',
        "encoding": "latin"
    },
    {
        "table_name": "stores",
        "columns": """StoreKey INT, Country VARCHAR(255), State VARCHAR(255), 
                      Square_Meters INT, Open_Date DATE""",
        "file_path": r'C:\Users\praba\Desktop\dataspark_project\df_files\df6_stores.csv',
        "encoding": "latin"
    }
]

# Create tables and insert data
for table in tables_and_files:
    # Create table
    create_table_query = f"""
    CREATE TABLE IF NOT EXISTS {table['table_name']} ({table['columns']});
    """
    cursor.execute(create_table_query)
    print(f"Table `{table['table_name']}` created successfully!")

    # Load data from CSV
    df = pd.read_csv(table['file_path'], encoding=table['encoding'])

    # Insert data into the table
    for _, row in df.iterrows():
        placeholders = ", ".join(["%s"] * len(row))
        columns = ", ".join([f"`{col}`" for col in df.columns])
        insert_query = f"""
        INSERT INTO {table['table_name']} ({columns}) VALUES ({placeholders});
        """
        cursor.execute(insert_query, tuple(row))
    connection.commit()
    print(f"Data from `{table['file_path']}` inserted into `{table['table_name']}` successfully!")

# Close the connection
cursor.close()
connection.close()
print("All tasks completed and database connection closed.")


```
